```python
from keras.applications.vgg16 import VGG16

model = VGG16() # Création du modèle VGG-16 implementé par Keras
```

```python
from keras.preprocessing.image import load_img, img_to_array
from keras.applications.vgg16 import preprocess_input

img1 = load_img('./images_test/image1.jpg', target_size=(224, 224))  # Charger l'image
img1 = img_to_array(img1)  # Convertir en tableau numpy
img1 = img1.reshape((1, img1.shape[0], img1.shape[1], img1.shape[2]))  # Créer la collection d'images (un seul échantillon)
img1 = preprocess_input(img1)  # Prétraiter l'image comme le veut VGG-16

img2 = load_img('./images_test/image2.jpg', target_size=(224, 224))  # Charger l'image
img2 = img_to_array(img2)  # Convertir en tableau numpy
img2 = img2.reshape((1, img2.shape[0], img2.shape[1], img2.shape[2]))  # Créer la collection d'images (un seul échantillon)
img2 = preprocess_input(img2)  # Prétraiter l'image comme le veut VGG-16

img3 = load_img('./images_test/image3.jpeg', target_size=(224, 224))  # Charger l'image
img3 = img_to_array(img3)  # Convertir en tableau numpy
img3 = img3.reshape((1, img3.shape[0], img3.shape[1], img3.shape[2]))  # Créer la collection d'images (un seul échantillon)
img3 = preprocess_input(img3)  # Prétraiter l'image comme le veut VGG-16
```

```python
y1 = model.predict(img1)  # Prédir la classe de l'image (parmi les 1000 classes d'ImageNet)
y2 = model.predict(img2)  # Prédir la classe de l'image (parmi les 1000 classes d'ImageNet)
y3 = model.predict(img3)  # Prédir la classe de l'image (parmi les 1000 classes d'ImageNet)
```

```python
from keras.applications.vgg16 import decode_predictions

# Afficher les 3 classes les plus probables
print('Top 3 :', decode_predictions(y1, top=3)[0])
print('Top 3 :', decode_predictions(y2, top=3)[0])
print('Top 3 :', decode_predictions(y3, top=3)[0])
```

### TRANSFERT LEARNING 

```python
# Packages 
import numpy as np 
import matplotlib.pyplot as plt 
from keras.layers import Dense,Conv2D,MaxPool2D,Flatten,SpatialDropout2D,Dropout
from keras.models import Sequential,Model
import time
from sklearn.model_selection import train_test_split
from keras.callbacks import TensorBoard,EarlyStopping
```

```python
# importing skin cancer dataset 
X_train = np.load('./archive/train_dataset/x_train.npy') 
X_test = np.load('./archive/test_dataset/x_test.npy')
y_train = np.load('./archive/train_dataset/y_train.npy')
y_test = np.load('./archive/test_dataset/y_test.npy')
```

```python
# plotting 6 firsts images 

plt.figure(figsize=(10,6))

plt.subplot(2, 3, 1)
plt.imshow(X_train[0])

plt.subplot(2, 3, 2)
plt.imshow(X_train[1])

plt.subplot(2, 3, 3)
plt.imshow(X_train[2])

plt.subplot(2, 3, 4)
plt.imshow(X_train[3])

plt.subplot(2, 3, 5)
plt.imshow(X_train[4])

plt.subplot(2, 3, 6)
plt.imshow(X_train[5])

plt.show()
```

```python
# Normalizing way to normalize 
X_train_std = X_train / abs(X_train.max()) 
X_test_std = X_test / abs(X_test.max())
```

```python
# splitting data 
x_train,x_val,y_train,y_val = train_test_split(X_train_std,y_train,test_size=0.1)
```

```python
# reshaping for VGG-16 
x_train_vgg = np.repeat(x_train,repeats=3,axis=3)
x_test_vgg = np.repeat(X_test_std,repeats=3,axis=3)
x_val_vgg = np.repeat(x_val,repeats=3,axis=3)
```

```python
# Importing VGG-16 model with no top 
vgg_model = VGG16(include_top = False,classes=2,input_shape=(128,128,3))
```

```python
vgg_model.summary()
```

```python
# Modifying model by adding 2 layers 
x = vgg_model.output
flatten_x = Flatten()(x)
d_1 = Dense(64,activation='relu')(flatten_x)
predictions = Dense(2, activation='softmax')(d_1)
```

```python
# Redifyning model 
vgg_model2 = Model(inputs=vgg_model.input, outputs=predictions)
NAME = f'vgg_model2_{time.time()}'
```

```python
# Training only new layers 
for layer in vgg_model2.layers[:19]:
    layer.trainable = False
```

```python
vgg_model2.summary()
```

```python
# Compiling and fitting model 

early = EarlyStopping(monitor='val_loss',patience=3)
vgg_model2.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

history = vgg_model2.fit(x_train_vgg,
                    y_train,
                    epochs=20,
                    batch_size=100,
                    validation_data=(x_val_vgg,y_val),
                    callbacks=[early])
```

```python
# Showing results 

val_acc = history.history['val_accuracy']
train_acc = history.history['accuracy']
fig,ax = plt.subplots(figsize=(14,7))
ax.plot(val_acc,label='validation accuracy')
ax.plot(train_acc,label='training accuracy')
ax.legend()
ax.set_title('Accuracy in train and validation set')
plt.show()

_, test_accuracy = vgg_model2.evaluate(x_test_vgg, y_test)
print(test_accuracy)
```
