### Data 

```python
# importing data 
X_train = np.load('./train_dataset/x_train.npy') 
X_test = np.load('./test_dataset/x_test.npy')
y_train = np.load('./train_dataset/y_train.npy')
y_test = np.load('./test_dataset/y_test.npy')
```

```python
# Normalizing way to normalize 
X_train = X_train / abs(X_train.max()) 
X_test = X_test / abs(X_test.max())
```

### Model 

```python
# Model architecture 

model = Sequential()
model.add(Conv2D(filters=64, kernel_size=2, padding='same', activation='relu', input_shape = X_train.shape[1:])) 
model.add(MaxPooling2D(pool_size=2))
model.add(Flatten())
model.add(Dense(2, activation='softmax'))
```

```python
# Model compilation 

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()
```

```python
# Fittting model 

model.fit(X_train, y_train, epochs=20, validation_split=0.1)
```

```python
# Testing anther model 

model1 = Sequential(
    [
        Input(shape=X_train.shape[1:]),
        Conv2D(32, kernel_size=(3, 3), activation="relu"),
        MaxPooling2D(pool_size=(2, 2)),
        Conv2D(64, kernel_size=(3, 3), activation="relu"),
        MaxPooling2D(pool_size=(2, 2)),
        Flatten(),  
        Dense(2, activation="softmax"),
    ]
)

model1.summary()
```

```python
batch_size = 128
epochs = 20

model1.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

model1.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, validation_split=0.1)
```

```python
# Testing anther model : introducing dropout  
# layers.Dropout(0.5), ## it is used during trainning to prevent overfitting by using only .5 of neurons. On l'utilise aprs une couche de neuronnes 
# https://lesdieuxducode.com/blog/2019/1/prototyper-un-reseau-de-neurones-avec-keras

model2 = Sequential(
    [
        Input(shape=X_train.shape[1:]),
        Conv2D(32, kernel_size=(3, 3), activation="relu"),
        # Dropout(0.2),
        MaxPooling2D(pool_size=(2, 2)),
        Conv2D(64, kernel_size=(3, 3), activation="relu"),
        # Dropout(0.2),
        MaxPooling2D(pool_size=(2, 2)),
        Flatten(),
        Dense(32, activation="relu"),
        Dropout(0.4),
        Dense(2, activation="softmax"),
    ]
)

model2.summary()
```

```python
# batch_size = 100
# epochs = 20

model2.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

# model1.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, validation_split=0.1)

# using early stopping 

early_stopping = EarlyStopping(monitor='val_accuracy', patience=5, verbose=1)

history = model2.fit(X_train, y_train, epochs=500, batch_size=100, validation_split=0.1, callbacks=[early_stopping])
```

### Data augmentation 

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
X_train2 = np.load('./train_dataset/x_train.npy')
y_train2 = np.load('./train_dataset/y_train.npy')

X_test2 = np.load('./test_dataset/x_test.npy')
y_test2 = np.load('./test_dataset/y_test.npy')

X_train2, X_val2, y_train2, y_val2 = train_test_split(X_train2, y_train2, test_size=0.1, random_state=15, stratify=y_train2)
```

```python
datagen = ImageDataGenerator(
    featurewise_center=True,
    # featurewise_std_normalization=True,
    rotation_range=180,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    # validation_split=0.1
    )

datagen.fit(X_train2)
```

```python
model3 = Sequential()
model3.add(Conv2D(filters=10, kernel_size=5, padding='same', activation='relu', input_shape = X_train2.shape[1:])) 
model3.add(MaxPooling2D(pool_size=2))
model3.add(Conv2D(filters=10, kernel_size=5, padding='same', activation='relu')) 
model3.add(MaxPooling2D(pool_size=2))
# model3.add(Conv2D(filters=30, kernel_size=5, padding='same', activation='relu', input_shape = X_train2.shape[1:])) 
# model3.add(MaxPooling2D(pool_size=2))
model3.add(Flatten())
model3.add(Dense(16, activation="relu"))
model3.add(Dense(10, activation="relu"))
model3.add(Dense(2, activation='softmax'))

# compiling 
model3.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
early_stopping3 = EarlyStopping(monitor='val_accuracy', patience=5, verbose=1)
history3 = model3.fit_generator(datagen.flow(X_train2, y_train2, batch_size=32), 
                    epochs=40, 
                    # batch_size =32, 
                    validation_data=(X_val2, y_val2),
                    callbacks=[early_stopping3],
                    )
```

```python
# evaluation  model 
test_loss3, test_acc3 = model3.evaluate(X_test2, y_test2)
print(test_loss3, test_acc3)
```
